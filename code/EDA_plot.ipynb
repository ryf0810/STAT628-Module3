```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
eda_data = pd.read_csv("EDA.csv")
correlation_matrix = eda_data[['review_count','stars_review', 'comment_length', 'AverageWordLength']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig('Correlation Matrix.png', format='png', dpi=100)
```

```python
sns.scatterplot(x='comment_length', y='stars_review', data=eda_data)
plt.title('Rating Stars vs Review Length')
plt.savefig('Rating Stars vs Review Length', format='png', dpi=300)
```
